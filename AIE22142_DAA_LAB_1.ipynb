1. Find the sum of first N natural numbers using Iterative and Recursive algorithms. Find the time taken to execute the same by varying ‘N’s value and plot it using python’s plot function.


```python
import time
import matplotlib.pyplot as plt


def recursive(n):
  result = 0
  while n > 0:
    result += n
    n -= 1
  return result




def iterative(n):
  sum = 0

  for i in range(1,n+1):
    sum += i
  return sum

N_values = [10, 100, 1000, 10000, 100000]
iterat_times = []
recur_times = []

for N in N_values:
    start_time = time.time()
    iterative(N)
    iterat_times.append(time.time() - start_time)

    start_time = time.time()
    recursive(N)
    recur_times.append(time.time() - start_time)

# Plotting the results
plt.plot(N_values, iterat_times, label='Iterative')
plt.plot(N_values, recur_times, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Time taken to find sum of first N natural numbers')
plt.legend()
plt.show()



```


2. Perform linear and binary searches for an array of 10000 elements. Use random function in Python to generate the integer array elements in the range 1 to 1000. The search key is an input given by the user. Plot the time taken by the algorithm for 5 different searches when executing the two algorithms.

```python
import time
import random
import matplotlib.pyplot as plt


def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1


def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Generating a random array consisting of 10000 elements
random_array = [random.randint(1, 1000) for _ in range(10000)]
random_array.sort()  # Sort the array for binary search

# Performing searches and plotting the time taken respectively
search_keys = [random.randint(1, 1000) for _ in range(5)]

linear_times = []
binary_times = []

for key in search_keys:
    start_time = time.time()
    linear_search(random_array, key)
    linear_times.append(time.time() - start_time)

    start_time = time.time()
    binary_search(random_array, key)
    binary_times.append(time.time() - start_time)

# Plotting the results
plt.plot(range(1, 6), linear_times, label='Linear Search')
plt.plot(range(1, 6), binary_times, label='Binary Search')
plt.xlabel('Search')
plt.ylabel('Time (seconds)')
plt.title('Time taken for linear and binary searches')
plt.legend()
plt.show()
```

3. Write a recursive function to convert the entered string of digits into the integer it represents. For example, 13531 represents the integer 13,531.

```python
def string_to_integer(s):
    if len(s) == 0:
        return 0

    first_digit = int(s[0])
    remaining_string = s[1:]
    integer_value = string_to_integer(remaining_string)

    result = first_digit * (10 ** len(remaining_string)) + integer_value

    return result


input_string = "13531"
result = string_to_integer(input_string)
formatted_result = "{:,}".format(result)  # Using format() function for formatting
print("Integer representation of '{}' is: {}".format(input_string, formatted_result))
```

```python
def reverse_string(s):
    if len(s) <= 1:
        return s
    else:
        return reverse_string(s[1:]) + s[0]


input_string = "ThisIsSparta"
reversed_string = reverse_string(input_string)
print("The reverse of '{}' is: {}".format(input_string, reversed_string))
```

Write a short recursive Python function that determines if a string s is a palindrome. For example, racecar and gohangasalamiimalasagnahog are palindromes.



```python
def is_palindrome(s):
    # Base case: if the length of the string is 0 or 1, it's a palindrome by default
    if len(s) <= 1:
        return True

    # Checking if the first and last characters are the same in the given string
    if s[0] != s[-1]:
        return False

    # Recursively check if the substring excluding the first and last characters is a palindrome
    return is_palindrome(s[1:-1])

# Example usage:
input_string1 = "BOB"
input_string2 = "frequecysoscillatingfrequency"

print("Is '{}' a palindrome? {}".format(input_string1, is_palindrome(input_string1)))
print("Is '{}' a palindrome? {}".format(input_string2, is_palindrome(input_string2)))

```
